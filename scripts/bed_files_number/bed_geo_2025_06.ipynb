```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("geo2025_06.csv")
```

```python
data
```

```python
data["summary"] = data["count"].cumsum()
```

```python
data["year_str"] = data["year"].astype(str)
```

```python
data
```

```python
fig, ax = plt.subplots()
plt.xticks(rotation=45)
ax.bar(data["year_str"], 
       data["summary"], 
       label=data["year_str"], 
       color="green")

ax.set_xlabel('Year')
ax.set_ylabel('Number of files')
ax.set_title('Cumulative number of BED files')
```

```python
fig.savefig('./bed_geo_2025_06_summary.svg')
```

```python
fig, ax = plt.subplots()
plt.xticks(rotation=45)
ax.bar(data["year_str"], 
       data["count"], 
       label=data["year_str"], 
       color="green")

ax.set_xlabel('Year')
ax.set_ylabel('Number of files')
ax.set_title('Number of BED files')
```

```python
fig.savefig('./bed_geo_2025_06_by_year.svg')
```
